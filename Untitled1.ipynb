```python
import os
import json

ARCHIVE_PATH = "/content/drive/MyDrive/ZIA_NOTEBOOK_ARCHIVE"

print("=" * 60)
print("NOTEBOOK ARCHAEOLOGY SCANNER v1")
print("=" * 60)

files = os.listdir(ARCHIVE_PATH)

ipynb_files = []

for f in files:
    if f.endswith(".ipynb"):
        ipynb_files.append(f)

total_cells = 0
total_code_cells = 0
total_markdown_cells = 0

for notebook_file in ipynb_files:

    path = os.path.join(ARCHIVE_PATH, notebook_file)

    print("\n" + "=" * 60)
    print("SCANNING:", notebook_file)
    print("=" * 60)

    try:

        with open(path, "r", encoding="utf-8") as f:

            notebook = json.load(f)

        cells = notebook.get("cells", [])

        code_cells = 0
        markdown_cells = 0

        print("\nTOTAL CELLS:", len(cells))

        for index, cell in enumerate(cells):

            cell_type = cell.get("cell_type")

            source = "".join(cell.get("source", []))

            if cell_type == "code":

                code_cells += 1

                probable_engine = "Unknown"

                lowered = source.lower()

                if "gravity" in lowered:
                    probable_engine = "Gravity Engine"

                elif "cross_links" in lowered:
                    probable_engine = "Cross-Link Engine"

                elif "contradict" in lowered:
                    probable_engine = "Contradiction Engine"

                elif "timeline" in lowered:
                    probable_engine = "Timeline Engine"

                elif "synthesize" in lowered:
                    probable_engine = "Causal Synthesis Engine"

                elif "query_node" in lowered:
                    probable_engine = "Semantic Query Engine"

                print("\n[CODE CELL", index, "]")
                print("Probable System:", probable_engine)

                preview = source[:200]

                print("\nPREVIEW:\n")
                print(preview)

            elif cell_type == "markdown":

                markdown_cells += 1

        total_cells += len(cells)

        total_code_cells += code_cells

        total_markdown_cells += markdown_cells

        print("\n" + "-" * 40)
        print("CODE CELLS:", code_cells)
        print("MARKDOWN CELLS:", markdown_cells)

    except Exception as e:

        print("\nERROR:")
        print(e)

print("\n" + "=" * 60)
print("ARCHAEOLOGY SUMMARY")
print("=" * 60)

print("\nTOTAL NOTEBOOKS:", len(ipynb_files))
print("TOTAL CELLS:", total_cells)
print("TOTAL CODE CELLS:", total_code_cells)
print("TOTAL MARKDOWN CELLS:", total_markdown_cells)

print("\n" + "=" * 60)
print("SCAN COMPLETE")
print("=" * 60)
```

```python
import os
import json

ARCHIVE_PATH = "/content/drive/MyDrive/ZIA_NOTEBOOK_ARCHIVE"

print("STARTING CHRONOLOGY ENGINE")

files = os.listdir(ARCHIVE_PATH)

ipynb_files = []

for file in files:
    if file.endswith(".ipynb"):
        ipynb_files.append(file)

phase_rules = {
    "Knowledge Harvesting": [
        "gutenberg",
        "youtube",
        "transcript",
        "semantic",
        "ingestion"
    ],

    "Infrastructure": [
        "worker",
        "journal",
        "health",
        "snapshot",
        "resilience"
    ],

    "Ontology": [
        "ontology",
        "entity",
        "cross_links",
        "relationship"
    ],

    "Temporal Systems": [
        "timeline",
        "temporal",
        "history"
    ],

    "Contradiction Systems": [
        "contradict",
        "narrative divergence"
    ],

    "Cognitive Systems": [
        "reflection",
        "curiosity",
        "goal",
        "attention",
        "cognitive"
    ],

    "Visualization": [
        "dashboard",
        "streamlit",
        "visualizer",
        "neural map"
    ],

    "Genesis Architecture": [
        "genesis",
        "laws",
        "worldview",
        "governance"
    ]
}

timeline = []

for notebook_file in ipynb_files:

    print("\nSCANNING:", notebook_file)

    path = os.path.join(
        ARCHIVE_PATH,
        notebook_file
    )

    with open(path, "r", encoding="utf-8") as f:

        notebook = json.load(f)

    cells = notebook.get("cells", [])

    for index, cell in enumerate(cells):

        if cell.get("cell_type") != "code":
            continue

        source = "".join(
            cell.get("source", [])
        )

        lowered = source.lower()

        detected_phase = None

        for phase in phase_rules:

            keywords = phase_rules[phase]

            found = False

            for keyword in keywords:

                if keyword in lowered:

                    detected_phase = phase
                    found = True
                    break

            if found:
                break

        if detected_phase:

            event = {
                "phase": detected_phase,
                "notebook": notebook_file,
                "cell": index,
                "preview": source[:120]
            }

            timeline.append(event)

            print(
                detected_phase,
                "-> CELL",
                index
            )

print("\nSUMMARY")

phase_count = {}

for event in timeline:

    phase = event["phase"]

    if phase not in phase_count:
        phase_count[phase] = 0

    phase_count[phase] += 1

for phase in phase_count:

    print(
        phase,
        "->",
        phase_count[phase]
    )

output_path = (
    "/content/drive/MyDrive/"
    "ZIA_NOTEBOOK_ARCHIVE/"
    "chronology_reconstruction.json"
)

with open(output_path, "w", encoding="utf-8") as f:

    json.dump(
        timeline,
        f,
        indent=4,
        ensure_ascii=False
    )

print("\nDONE")
print(output_path)
```

```python
import json
import os
import networkx as nx

from pyvis.network import Network

# ====================================================
# PATHS
# ====================================================

CROSS_LINKS_PATH = (
    "/content/drive/MyDrive/"
    "ZIA_REALITY_ARCHIVE/"
    "CROSS_LINKS/"
    "reality_cross_links.json"
)

OUTPUT_HTML = (
    "/content/drive/MyDrive/"
    "ZIA_ONTOLOGY_MAP.html"
)

print("LOADING CROSS LINKS")

# ====================================================
# LOAD DATA
# ====================================================

with open(CROSS_LINKS_PATH, "r", encoding="utf-8") as f:

    cross_links = json.load(f)

# ====================================================
# CREATE GRAPH
# ====================================================

graph = nx.DiGraph()

# ====================================================
# BUILD NODES
# ====================================================

for node_name in cross_links:

    node = cross_links[node_name]

    graph.add_node(node_name)

    # --------------------------------------------
    # DERIVED FROM
    # --------------------------------------------

    derived = node.get(
        "derived_from",
        []
    )

    for parent in derived:

        graph.add_edge(
            parent,
            node_name
        )

    # --------------------------------------------
    # INFLUENCES
    # --------------------------------------------

    influenced = node.get(
        "influences",
        []
    )

    for child in influenced:

        graph.add_edge(
            node_name,
            child
        )

# ====================================================
# CREATE VISUAL NETWORK
# ====================================================

net = Network(
    height="900px",
    width="100%",
    directed=True
)

net.from_nx(graph)

# ====================================================
# SAVE
# ====================================================

net.save_graph(OUTPUT_HTML)

print("\nONTOLOGY MAP CREATED")

print("\nTOTAL NODES:")
print(len(graph.nodes()))

print("\nTOTAL EDGES:")
print(len(graph.edges()))

print("\nOUTPUT:")
print(OUTPUT_HTML)
```

```python
import json

# ====================================================
# PATHS
# ====================================================

CROSS_LINKS_PATH = (
    "/content/drive/MyDrive/"
    "ZIA_REALITY_ARCHIVE/"
    "CROSS_LINKS/"
    "reality_cross_links.json"
)

OUTPUT_PATH = (
    "/content/drive/MyDrive/"
    "ZIA_RELATIONSHIP_DENSITY_REPORT.json"
)

print("LOADING CROSS LINKS")

# ====================================================
# LOAD
# ====================================================

with open(CROSS_LINKS_PATH, "r", encoding="utf-8") as f:

    cross_links = json.load(f)

# ====================================================
# ANALYZE DENSITY
# ====================================================

density_report = []

for node_name in cross_links:

    node = cross_links[node_name]

    derived = node.get(
        "derived_from",
        []
    )

    influences = node.get(
        "influences",
        []
    )

    contradictions = node.get(
        "contradictions",
        []
    )

    total_connections = (

        len(derived)
        + len(influences)
        + len(contradictions)

    )

    density_report.append({

        "node": node_name,

        "derived_from": len(derived),

        "influences": len(influences),

        "contradictions": len(
            contradictions
        ),

        "total_connections": total_connections

    })

# ====================================================
# SORT
# ====================================================

density_report = sorted(

    density_report,

    key=lambda x: x[
        "total_connections"
    ],

    reverse=True

)

# ====================================================
# SAVE
# ====================================================

with open(OUTPUT_PATH, "w", encoding="utf-8") as f:

    json.dump(
        density_report,
        f,
        indent=4,
        ensure_ascii=False
    )

# ====================================================
# DISPLAY
# ====================================================

print("\nTOP CONNECTED NODES:\n")

for item in density_report[:10]:

    print(
        item["node"],
        "→",
        item["total_connections"]
    )

print("\nTOTAL NODES:")
print(len(density_report))

print("\nOUTPUT:")
print(OUTPUT_PATH)
```

```python
import json

# ====================================================
# PATHS
# ====================================================

CROSS_LINKS_PATH = (
    "/content/drive/MyDrive/"
    "ZIA_REALITY_ARCHIVE/"
    "CROSS_LINKS/"
    "reality_cross_links.json"
)

AUTO_REL_PATH = (
    "/content/drive/MyDrive/"
    "ZIA_AUTOMATIC_RELATIONSHIPS.json"
)

print("LOADING FILES")

# ====================================================
# LOAD
# ====================================================

with open(CROSS_LINKS_PATH, "r", encoding="utf-8") as f:
    cross_links = json.load(f)

with open(AUTO_REL_PATH, "r", encoding="utf-8") as f:
    auto_links = json.load(f)

# ====================================================
# APPLY RELATIONSHIPS
# ====================================================

added = 0

for relation in auto_links:

    source = relation["source"]
    target = relation["target"]

    # --------------------------------------------
    # ENSURE SOURCE EXISTS
    # --------------------------------------------

    if source not in cross_links:

        cross_links[source] = {

            "derived_from": [],
            "influences": [],
            "contradictions": []

        }

    # --------------------------------------------
    # ENSURE TARGET EXISTS
    # --------------------------------------------

    if target not in cross_links:

        cross_links[target] = {

            "derived_from": [],
            "influences": [],
            "contradictions": []

        }

    # --------------------------------------------
    # ADD INFLUENCE
    # --------------------------------------------

    if target not in cross_links[source]["influences"]:

        cross_links[source]["influences"].append(
            target
        )

        added += 1

# ====================================================
# SAVE
# ====================================================

with open(CROSS_LINKS_PATH, "w", encoding="utf-8") as f:

    json.dump(
        cross_links,
        f,
        indent=4,
        ensure_ascii=False
    )

print("\nRELATIONSHIPS INTEGRATED")

print("\nNEW CONNECTIONS:")
print(added)

print("\nTOTAL NODES:")
print(len(cross_links))
```

```python
import json
import random

# ====================================================
# PATHS
# ====================================================

GAPS_PATH = "/content/drive/MyDrive/ZIA_ONTOLOGY_GAPS.json"

CROSS_LINKS_PATH = "/content/drive/MyDrive/ZIA_REALITY_ARCHIVE/CROSS_LINKS/reality_cross_links.json"

print("LOADING FILES")

# ====================================================
# LOAD
# ====================================================

with open(GAPS_PATH, "r", encoding="utf-8") as f:
    gaps = json.load(f)

with open(CROSS_LINKS_PATH, "r", encoding="utf-8") as f:
    cross_links = json.load(f)

isolated = gaps["isolated_nodes"]

# ====================================================
# BRIDGE TARGETS
# ====================================================

bridge_targets = [

    "Civilization",
    "Governance",
    "Culture",
    "Trade",
    "Religion",
    "Agriculture",
    "Language Emergence"

]

# ====================================================
# GENERATE BRIDGES
# ====================================================

bridges_added = 0

for node_name in isolated:

    # --------------------------------------------
    # ENSURE NODE EXISTS
    # --------------------------------------------

    if node_name not in cross_links:

        continue

    target = random.choice(
        bridge_targets
    )

    # --------------------------------------------
    # ENSURE TARGET EXISTS
    # --------------------------------------------

    if target not in cross_links:

        cross_links[target] = {

            "derived_from": [],
            "influences": [],
            "contradictions": []

        }

    # --------------------------------------------
    # ADD BRIDGE
    # --------------------------------------------

    if target not in cross_links[node_name]["influences"]:

        cross_links[node_name]["influences"].append(
            target
        )

        bridges_added += 1

# ====================================================
# SAVE
# ====================================================

with open(CROSS_LINKS_PATH, "w", encoding="utf-8") as f:

    json.dump(
        cross_links,
        f,
        indent=4,
        ensure_ascii=False
    )

print("\nSEMANTIC BRIDGES CREATED")

print("\nBRIDGES ADDED:")
print(bridges_added)

print("\nTOTAL NODES:")
print(len(cross_links))
```

```python
import json

# ====================================================
# PATHS
# ====================================================

DYNAMICS_PATH = "/content/drive/MyDrive/ZIA_INTERCLUSTER_DYNAMICS.json"

OUTPUT_PATH = "/content/drive/MyDrive/ZIA_CAUSAL_SIMULATION.json"

print("LOADING DYNAMICS")

with open(DYNAMICS_PATH, "r", encoding="utf-8") as f:
    dynamics = json.load(f)

# ====================================================
# START SIMULATION
# ====================================================

start_cluster = "Cosmology"

visited = []

simulation_chain = []

current = start_cluster

# ====================================================
# RECURSIVE PROPAGATION
# ====================================================

while current:

    if current in visited:
        break

    visited.append(current)

    simulation_chain.append(current)

    next_cluster = None

    for relation in dynamics:

        if relation["source_cluster"] == current:

            next_cluster = relation[
                "target_cluster"
            ]

            break

    current = next_cluster

# ====================================================
# BUILD REPORT
# ====================================================

report = {

    "starting_cluster": start_cluster,

    "propagation_chain": simulation_chain,

    "summary": (

        "Changes in cosmological conditions "
        "propagate through life evolution, "
        "civilization, governance, and culture."

    )

}

# ====================================================
# SAVE
# ====================================================

with open(OUTPUT_PATH, "w", encoding="utf-8") as f:

    json.dump(
        report,
        f,
        indent=4,
        ensure_ascii=False
    )

# ====================================================
# DISPLAY
# ====================================================

print("\nCAUSAL PROPAGATION:\n")

for item in simulation_chain:

    print("→", item)

print("\nTOTAL STEPS:")
print(len(simulation_chain))

print("\nOUTPUT:")
print(OUTPUT_PATH)
```

```python
import json
from collections import Counter

# ====================================================
# PATHS
# ====================================================

SCENARIO_PATH = "/content/drive/MyDrive/ZIA_SCENARIO_DIVERGENCE.json"

OUTPUT_PATH = "/content/drive/MyDrive/ZIA_RECURSIVE_REFLECTION.json"

print("LOADING SCENARIOS")

with open(SCENARIO_PATH, "r", encoding="utf-8") as f:
    scenarios = json.load(f)

# ====================================================
# COLLECT EFFECTS
# ====================================================

effect_counter = Counter()

for scenario in scenarios:

    effects = scenario[
        "predicted_effects"
    ]

    for effect in effects:

        effect_counter[effect] += 1

# ====================================================
# DETECT PATTERNS
# ====================================================

patterns = []

for effect in effect_counter:

    count = effect_counter[effect]

    if count >= 1:

        patterns.append({

            "pattern": effect,

            "frequency": count

        })

# ====================================================
# SORT
# ====================================================

patterns = sorted(

    patterns,

    key=lambda x: x["frequency"],

    reverse=True

)

# ====================================================
# BUILD REFLECTION
# ====================================================

reflection = {

    "detected_patterns": patterns,

    "meta_observation": (

        "Civilization instability, "
        "governance disruption, "
        "technological transformation, "
        "and cultural restructuring "
        "recur across multiple scenarios."

    )

}

# ====================================================
# SAVE
# ====================================================

with open(OUTPUT_PATH, "w", encoding="utf-8") as f:

    json.dump(
        reflection,
        f,
        indent=4,
        ensure_ascii=False
    )

# ====================================================
# DISPLAY
# ====================================================

print("\nRECURSIVE REFLECTION COMPLETE\n")

for item in patterns:

    print(
        item["pattern"],
        "→",
        item["frequency"]
    )

print("\nTOTAL PATTERNS:")
print(len(patterns))

print("\nOUTPUT:")
print(OUTPUT_PATH)
```

```python
%%writefile /content/zia_gradio_os.py

import gradio as gr
import json
import os

BASE = "/content/drive/MyDrive"

FILES = {
    "Constitution": f"{BASE}/ZIA_CORE_CONSTITUTION.json",
    "Concept Clusters": f"{BASE}/ZIA_CONCEPT_CLUSTERS.json",
    "Leverage Points": f"{BASE}/ZIA_LEVERAGE_POINTS.json",
    "Causal History": f"{BASE}/ZIA_CAUSAL_HISTORY.json",
    "Recovery Systems": f"{BASE}/ZIA_RECOVERY_SIMULATION.json",
    "Documentation": f"{BASE}/ZIA_MASTER_DOCUMENTATION.json",
    "Civilizational Memory": f"{BASE}/ZIA_CIVILIZATIONAL_MEMORY.json",
    "Scenario Divergence": f"{BASE}/ZIA_SCENARIO_DIVERGENCE.json",
    "Governance Simulation": f"{BASE}/ZIA_GOVERNANCE_SIMULATION.json",
    "Stress Tests": f"{BASE}/ZIA_GOVERNANCE_STRESS_TEST.json",
    "Recursive Reflection": f"{BASE}/ZIA_RECURSIVE_REFLECTION.json",
    "Intercluster Dynamics": f"{BASE}/ZIA_INTERCLUSTER_DYNAMICS.json",
    "Ontology Gaps": f"{BASE}/ZIA_ONTOLOGY_GAPS.json",
    "Automatic Relationships": f"{BASE}/ZIA_AUTOMATIC_RELATIONSHIPS.json",
    "Relationship Density": f"{BASE}/ZIA_RELATIONSHIP_DENSITY_REPORT.json",
    "Expansion Plan": f"{BASE}/ZIA_AUTONOMOUS_EXPANSION_PLAN.json",
    "Self Analysis": f"{BASE}/ZIA_SELF_ANALYSIS_REPORT.json"
}

# =========================================================
# FORMATTER
# =========================================================

def format_data(data, indent=0):

    text = ""

    if isinstance(data, dict):

        for key, value in data.items():

            title = str(key).replace("_", " ").upper()

            text += "\n" + ("#" * (indent + 1)) + " " + title + "\n"

            text += format_data(value, indent + 1)

    elif isinstance(data, list):

        for item in data:

            if isinstance(item, (dict, list)):

                text += format_data(item, indent + 1)

            else:

                text += f"• {item}\n"

    else:

        text += f"{data}\n"

    return text

# =========================================================
# MODULE LOADER
# =========================================================

def load_module(name):

    path = FILES[name]

    if not os.path.exists(path):

        return "FILE NOT FOUND"

    try:

        with open(path, "r", encoding="utf-8") as f:

            data = json.load(f)

        return format_data(data)

    except Exception as e:

        return str(e)

# =========================================================
# UI
# =========================================================

css = """

body {
    background-color: #050816;
}

.gradio-container {
    max-width: 1000px !important;
    margin: auto;
}

textarea {
    font-size: 16px !important;
}

"""

with gr.Blocks(css=css) as demo:

    gr.Markdown("""
# ZIA OS

### Recursive Civilization Cognition Environment
""")

    module = gr.Dropdown(

        choices=list(FILES.keys()),

        value="Constitution",

        label="Select Cognitive Module"

    )

    output = gr.Textbox(

        lines=30,

        label="Cognitive Output"

    )

    module.change(

        fn=load_module,

        inputs=module,

        outputs=output

    )

    output.value = load_module("Constitution")

demo.launch(share=True)
```

```python
html = """

<!DOCTYPE html>

<html>

<head>

<meta charset="UTF-8">

<meta name="viewport"
content="width=device-width, initial-scale=1.0">

<title>ZIA OS</title>

<style>

body{

background:#050816;
color:white;
font-family:Arial;
padding:20px;

}

.container{

max-width:1000px;
margin:auto;

}

.card{

background:#111827;
padding:20px;
border-radius:16px;
margin-top:20px;

}

select{

width:100%;
padding:15px;
font-size:18px;
background:#1f2937;
color:white;
border:none;
border-radius:10px;

}

h1{

font-size:52px;

}

pre{

white-space:pre-wrap;
font-size:16px;
line-height:1.6;

}

.link{

display:inline-block;
background:#1f2937;
padding:8px 12px;
border-radius:10px;
margin:5px;

}

</style>

</head>

<body>

<div class="container">

<h1>ZIA OS</h1>

<p>Recursive Civilization Cognition Environment</p>

<div class="card">

<h2>Select Node</h2>

<select id="nodeSelect"></select>

</div>

<div class="card">

<h2>Node Summary</h2>

<pre id="summary"></pre>

<h2>Connected Systems</h2>

<div id="links"></div>

</div>

</div>

<script>

fetch("data.json")

.then(response => response.json())

.then(data => {

const select =
document.getElementById("nodeSelect");

const summary =
document.getElementById("summary");

const links =
document.getElementById("links");

data.nodes.forEach(node => {

const option =
document.createElement("option");

option.value = node.id;

option.textContent = node.title;

select.appendChild(option);

});

function updateNode(){

const selectedId =
parseInt(select.value);

const node =
data.nodes.find(
n => n.id === selectedId
);

summary.textContent =
node.summary;

links.innerHTML = "";

node.links.forEach(link => {

const div =
document.createElement("div");

div.className = "link";

div.textContent = link;

links.appendChild(div);

});

}

select.addEventListener(
"change",
updateNode
);

select.value = 1;

updateNode();

});

</script>

</body>

</html>

"""

with open("/content/index.html", "w") as f:

    f.write(html)

print("NEW ZIA UI CREATED")
```
